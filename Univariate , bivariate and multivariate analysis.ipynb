# Univariate , bivariate and multivariate analysis

in the univatiate analysis we try to pick up one feature and based on that feature we try to determine the outputs

in the Bivatiate analysis we try to pick up two features and based on that feature we try to determine the outputs

in multivariate analysis we take on the multiple features and the try to grout the output categories

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sns
```

```python
dataset = pd.read_csv("https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv")
```

# Univariate analysis

```python
SETOSA = dataset.loc[dataset['species']=='setosa']
VIRGINICA = dataset.loc[dataset['species']=='virgnica']
VERSICOLOR = dataset.loc[dataset['species']=='versicolor']
```

```python
SETOSA # it's displaying all the features where species == setosa
```

```python
plt.plot(SETOSA['sepal_length'],np.zeros_like(SETOSA['sepal_length']),'o')
plt.plot(VIRGINICA['sepal_length'],np.zeros_like(VIRGINICA['sepal_length']),'o')
plt.plot(VERSICOLOR['sepal_length'],np.zeros_like(VERSICOLOR['sepal_length']),'o')
plt.xlabel("SEPAL LENGTH")
plt.show()
```

# BIVARIATE ANALYSIS

```python
sns.FacetGrid(dataset,hue='species',size=5).map(plt.scatter,"sepal_length","sepal_width").add_legend();
```

```python
# add_legend() helps us to show what points have been plotted by which color
```

# MULTIVARIATE ANALYSIS
