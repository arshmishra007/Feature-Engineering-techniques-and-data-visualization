```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
y=np.arange(11,21)
```

```python
print(x)
```

```python
print(y)
```

```python
##Scatter plot
```

```python

plt.scatter(x,y,c='red')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Scatter Graph")
```

```python
#changing the input_data
y=x*x
```

```python
print(y)
```

```python
#linear curve
plt.plot(x,y)
```

```python
#for creating a little styled graph(use linestyle property)
plt.plot(x,y,'red',linestyle='dashed',linewidth=5)
```

```python
#creating the sine wave form
pi=np.pi
print(pi)
```

```python
x = np.arange(0,7*pi,0.4)
y=np.sin(x)
```

```python
plt.plot(x,y)
```

```python
##Barplot 
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
plt.bar(x,y)
plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.title("Barplot")
```

```python
##Histogram
```

```python
a=np.array([2,3,4,5,12,31,13,45,66,43])
plt.hist(a)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Histogram")
```

```python
##piechart
```

```python
labels = 'chocolate','chicken','paneer','indian sweets'
sizes=[100,250,250,200]
colors=['brown','red','green','yellow']
explode=(0,0.2,0.1,0)
plt.pie(sizes,explode=explode,labels=labels,colors=colors,shadow=True)
```
